```python
beras = 11400
minyak = 12700
ayam = 3400
telur = 21800
popok = 27000
sabun_cuci = 1800
buah_anggur = 3800
roti_tawar = 8500
tissu = 6400
mie_instan = 2200
print("Belanjaan tia")
beras_tia = 4
ayam_tia = 2
telur_tia = 3
sabun_tia = 5
anggur_tia = 4
roti_tia = 2
mie_tia = 12

jb_tia = (beras * beras_tia) + (ayam * (ayam_tia * 10 )) + (telur * telur_tia) + (sabun_cuci * sabun_tia) + (buah_anggur * anggur_tia) + (roti_tawar * roti_tia) + (mie_instan * mie_tia)
print("Maka jumlah belanjaan tia adalah " + str(jb_tia))
diskon = 85
jb_diskon_tia = (jb_tia / 100) * diskon
print("Maka jumlah belanjaan diskon tia adalah " + str(jb_diskon_tia))

print("")
print("Belanjaan Hana")
minyak_hana = 2
ayam_hana = 5
telur_hana = 1.5
sabun_hana = 3
anggur_hana = 8
roti_hana = 3
tissu_hana = 2
mie_hana = 20

jb_hana = (minyak * minyak_hana) + (ayam * ayam_hana ) + (telur * telur_hana) + (sabun_cuci * sabun_hana) + (buah_anggur * anggur_hana) + (roti_tawar * roti_hana) + (tissu * tissu_hana) + (mie_instan * mie_hana)
print("Maka jumlah belanjaan Hana adalah " + str(jb_hana))
diskon = 90
jb_diskon_hana = (jb_hana / 100) * diskon
print("Maka jumlah belanjaan diskon tia adalah " + str(jb_diskon_hana))

print("")
print("Belanjaan Widia")
beras_widia = 1
minyak_widia = 1
telur_widia = 1
anggur_widia = 5
tissu_widia = 3
mie_widia = 16

jb_widia = (beras * beras_widia) + (minyak* minyak_widia ) + (telur * telur_widia) + (buah_anggur * anggur_widia) + (tissu * tissu_widia) + (mie_instan * mie_widia)
print("Maka jumlah belanjaan Widia adalah " + str(jb_widia))

print("")
print("Belanjaan Romi")
minyak_romi = 3
telur_romi = 2
popok_romi = 4
sabun_romi = 8
mie_romi = 8

jb_romi = (minyak* minyak_romi ) + (telur * telur_romi) + (popok * popok_romi) + (sabun_cuci * sabun_romi) + (mie_instan * mie_romi)
print("Maka jumlah belanjaan Romi adalah " + str(jb_romi))
diskon = 85
jb_diskon_romi = (jb_romi / 100) * diskon
print("Maka jumlah belanjaan diskon Romi adalah " + str(jb_diskon_romi))

print("")
print("Belanjaan Samsul")
ayam_samsul = 3
telur_samsul = 4
popok_samsul = 2
anggur_samsul = 7
tissu_samsul = 5
mie_samsul = 10

jb_samsul = (ayam * (ayam_samsul * 10)) + (telur * telur_samsul) + (popok * popok_samsul) + (buah_anggur * anggur_samsul) + (tissu * tissu_samsul) + (mie_instan * mie_samsul)
print("Maka jumlah belanjaan Samsul adalah " + str(jb_samsul))
diskon = 75
jb_diskon_samsul = (jb_samsul / 100) * diskon
print("Maka jumlah belanjaan diskon Samsul adalah " + str(jb_diskon_samsul))

print("")
print("Belanjaan Dimas")
ayam_dimas = 5
popok_dimas = 5
sabun_dimas = 4
anggur_dimas = 3
roti_dimas = 5
mie_dimas = 6

jb_dimas = (ayam * (ayam_dimas * 10)) + (popok * popok_dimas) + (sabun_cuci * sabun_dimas) + (buah_anggur * anggur_dimas) + (roti_tawar * roti_dimas) + (mie_instan * mie_dimas)
print("Maka jumlah belanjaan Dimas adalah " + str(jb_dimas))
diskon = 75
jb_diskon_dimas = (jb_dimas / 100) * diskon
print("Maka jumlah belanjaan diskon Dimas adalah " + str(jb_diskon_dimas))

print("")
max = [284475]
print("Maka Jumlah belanjaan paling besar setelah diskon adalah dimas Rp."+str(max))

print("")
tenpersen={"hana" : "Rp.173880"}
print("Maka yang mendapatkan diskon 10% Adalah: "+ str(tenpersen))
```

```python
#Pie chart
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
nama = ['Tia','Hana','Widia','Romi','Samsul','Dimas']
harga = [246600, 193200, 119300,221700,323800,379300]
ax.pie(harga,labels=nama,autopct="%1.2f%%")
plt.title("Keseluruhan Belanjaan")
plt.show()
```
